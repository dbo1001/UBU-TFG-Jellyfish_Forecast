```python
import os
import xarray as xr 
import pandas as pd
from math import floor
from datetime import datetime

def floorQuarter(x):
    return (floor(x * 4) / 4.0) 
```

```python
xr.set_options(display_style="html")
# lectura de archivo .nc
data = xr.open_dataset('../descargas/mercatorglorys12v1_gl12_mean_20150527_R20150603__filtrados.nc')
# xarray a dataFrame
dfC= data.to_dataframe()
# elimino los indices para modificar las fechas
dfC = dfC.reset_index()
# elimino la hora de la fechas
dfC['time'] = dfC['time'].dt.strftime('%Y-%m-%d')
# se redondea las coordnadas a un cuarto de grado para tener la misma precision que el dataframe de los avistamientos
dfC["latitude_red"] = dfC.latitude.map(floorQuarter)
dfC["longitude_red"] = dfC.longitude.map(floorQuarter)
# vuelvo a generar los indices
#dfC = dfC.set_index(["depth","latitude","longitude"])
dfC = dfC.set_index(["longitude","latitude"])
dfC
```
